```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
```

```python
scooter_df = pd.read_csv('https://raw.githubusercontent.com/iamzhenxp/E4418-Transportation-Anlytics-Logistic/FInal-project/FInal%20Project/Scooter%20Data.csv', encoding='unicode_escape')
centerline_df = pd.read_csv('https://raw.githubusercontent.com/iamzhenxp/E4418-Transportation-Anlytics-Logistic/FInal-project/FInal%20Project/CenterlineID.csv', encoding='unicode_escape')
ziplist = pd.read_csv('https://gist.githubusercontent.com/erichurst/7882666/raw/5bdc46db47d9515269ab12ed6fb2850377fd869e/US%2520Zip%2520Codes%2520from%25202013%2520Government%2520Data', encoding='unicode_escape')
```

```python
ziplist2= pd.read_csv('https://raw.githubusercontent.com/carstonhernke/dockless-scooter-analysis/master/scooter_trips_with_coordinates.csv', encoding='unicode_escape')
data_18 = pd.read_csv('https://raw.githubusercontent.com/carstonhernke/dockless-scooter-analysis/master/enriched_scooter_trips.csv', encoding='unicode_escape')
```

```python
centerline_zipcode = pd.DataFrame(centerline_df[['GBSID', 'ZIP5_L']])
```

```python
scooter_zipcode_df = pd.merge(scooter_df, centerline_zipcode, left_on='StartCenterlineID', right_on = 'GBSID', how="left")
scooter_zipcode_df = scooter_zipcode_df.rename(columns = {'ZIP5_L':'Zipcode'})
scooter_zipcode_df = pd.merge(scooter_zipcode_df, ziplist, left_on='Zipcode', right_on = 'ZIP', how="left")
```

```python
coordinate = pd.merge(ziplist2,data_18,left_on='TripID',right_on='TripID',how="left")
coordinate = coordinate[['StartCenterlineID','start_lat','start_lon']]
```

```python
scooter_zipcode_df = scooter_zipcode_df[['TripDuration', 'TripDistance', 'StartTime', 'StartCenterlineID', 'Zipcode','LAT','LNG']].dropna()
scooter_zipcode_df
```

```python
coordinate = coordinate.loc[coordinate.StartCenterlineID[scooter_zipcode_df['StartCenterlineID']].unique()]
coordinate
```

```python
plot_data = pd.merge(scooter_zipcode_df,coordinate,left_on='StartCenterlineID',right_on='StartCenterlineID',how="left")
plot_data = plot_data.dropna()
plot_data = plot_data.drop_duplicates()
plot_data.to_csv("Minneapolis_data_coordinate.csv")
```
